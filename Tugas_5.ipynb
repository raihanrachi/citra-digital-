```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Sampling frequency
fs = 500

# Membuat waktu
t = np.linspace(0, 1, fs, endpoint=False)

# Membuat sinyal gabungan (5Hz + 50Hz + 120Hz)
x = np.sin(2*np.pi*5*t) + np.sin(2*np.pi*50*t) + np.sin(2*np.pi*120*t)

# Fungsi untuk plot
def plot_filter_result(title, filtered_signal):
    plt.figure(figsize=(10,3))
    plt.plot(t, filtered_signal)
    plt.title(title)
    plt.grid()
    plt.show()

# --- 1. LOW PASS FILTER ---
sos_lp = signal.butter(10, 15, 'lp', fs=fs, output='sos')
low_passed = signal.sosfilt(sos_lp, x)
plot_filter_result("1. Low-Pass Filter (Frekuensi rendah <15Hz)", low_passed)

# --- 2. HIGH PASS FILTER ---
sos_hp = signal.butter(10, 100, 'hp', fs=fs, output='sos')
high_passed = signal.sosfilt(sos_hp, x)
plot_filter_result("2. High-Pass Filter (Frekuensi tinggi >100Hz)", high_passed)

# --- 3. BAND PASS FILTER ---
sos_bp = signal.butter(10, [40,60], 'bp', fs=fs, output='sos')
band_passed = signal.sosfilt(sos_bp, x)
plot_filter_result("3. Band-Pass Filter (Frekuensi 40-60Hz)", band_passed)
```
